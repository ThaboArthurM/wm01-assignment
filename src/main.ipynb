```python
df = pd.concat([df_za, df_les], ignore_index=True)
df
```

```python
df.to_csv("../data/cleaned/combined_learner_performance.csv")
```

```python
df["gender"] = df["gender"].map({'Male': 1, 'Female': 0})
df
```

```python
df["completion_status"] = df["completion_status"].map({"Completed": 1, "Not Completed": 0})

df
```

```python
gender_counts = df.gender.value_counts().rename({1: 'Male', 0: 'Female'})
gender_counts.columns = ["gender", "total"]
gender_counts
```

```python
completion_per = (df.groupby('gender')["completion_status"].mean().reset_index())

completion_per['completion_percent'] = completion_per['completion_status'] * 100
completion_per
```

```python
df["attendence_rate"] = (df["classes_attended"] / df["total_classes"] * 100).round()
df
```

```python
coun = df.groupby("country")["completion_status"].sum()
coun
```
